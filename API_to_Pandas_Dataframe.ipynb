```python
# Final Optimized Code
import pandas as pd
import requests

final_list = []

url = "https://api.themoviedb.org/3/trending/movie/day?language=en-US"

headers = {
    "accept": "application/json",
    "Authorization": "Bearer eyJhbGciOiJIUzI1NiJ9.eyJhdWQiOiIxNzJjOGI5M2ZlNTVlZjFiOWY0YWIxNDdmYzVkNGY3NyIsIm5iZiI6MTcyNDQyNTczMS4zNDE2OTYsInN1YiI6IjY2YzhhMDA0MTc0OGU4YmJlZjFkNDE1YyIsInNjb3BlcyI6WyJhcGlfcmVhZCJdLCJ2ZXJzaW9uIjoxfQ.a0VuHNTRcAUX6lm3cWGUCFl7hRxWCxsRtgxoJTDsoZ0"
}

try:
    response = requests.get(url, headers=headers)
    response.raise_for_status()
    response_json = response.json()
    total_pages = response_json.get('total_pages', 0)

except requests.exceptions.RequestException as e:
    print(f"Request failed: {e}")
    total_pages = 0
except KeyError as e:
    print(f"Key error: {e}")
    total_pages = 0

for i in range(1, total_pages + 1):
    try:
        url = f"https://api.themoviedb.org/3/trending/movie/day?language=en-US&page={i}"
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        response_json = response.json()
        results = response_json.get('results', [])

        title = []
        original_title = []
        overview = []
        original_language = []
        popularity = []
        release_date = []
        vote_average = []
        vote_count = []

        for movie in results:
            try:
                title.append(movie.get('title', None))
                original_title.append(movie.get('original_title', None))
                overview.append(movie.get('overview', None))
                original_language.append(movie.get('original_language', None))
                popularity.append(movie.get('popularity', None))
                release_date.append(movie.get('release_date', None))
                vote_average.append(movie.get('vote_average', None))
                vote_count.append(movie.get('vote_count', None))
            except Exception as e:
                print(f"Error processing movie: {e}")
                continue  # Skip to the next movie

        tmdb_dict = {
            'title': title,
            'original_title': original_title,
            'overview': overview,
            'original_language': original_language,
            'popularity': popularity,
            'release_date': release_date,
            'vote_average': vote_average,
            'vote_count': vote_count
        }

        df = pd.DataFrame(tmdb_dict)
        final_list.append(df)

    except requests.exceptions.RequestException as e:
        print(f"Request failed on page {i}: {e}")
    except KeyError as e:
        print(f"Key error on page {i}: {e}")
    except Exception as e:
        print(f"Unexpected error on page {i}: {e}")

if final_list:
    final = pd.concat(final_list, ignore_index=True)
else:
    print("No data to concatenate into the final DataFrame.")
```
