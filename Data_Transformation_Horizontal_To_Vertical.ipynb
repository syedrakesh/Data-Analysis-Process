```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('~/Desktop/ai_1.csv')

# Create an empty DataFrame for the transformed data
transformed_df = pd.DataFrame(columns=['Sl. No.', 'Product Name', 'Product Site', 'Review Generated By', 'Reviews'])

# Iterate over each row in the original DataFrame
for index, row in df.iterrows():
    # Extract the frozen columns
    serial_number = row['Sl. No.']
    product_name = row['Product Name']
    product_site = row['Product Site']
    review_generated_by = row['Review Generated By']
    
    # Iterate over the review columns that need to be melted into rows
    reviews = [
        'Review - 1', 'Review - 2', 'Review - 3', 'Review - 4', 'Review - 5', 
        'Review - 6', 'Review - 7', 'Review - 8', 'Review - 9', 'Review - 10', 
        'Review - 11', 'Review - 12', 'Review - 13', 'Review - 14', 'Review - 15', 
        'Review - 16', 'Review - 17', 'Review - 18', 'Review - 19', 'Review - 20'
    ]
    
    # Prepare a list to hold the rows to be concatenated
    rows_to_concat = []
    
    for review in reviews:
        # Check if the review is not empty
        if pd.notna(row[review]):
            # Create a new row as a DataFrame
            new_row = pd.DataFrame({
                'Sl. No.': [serial_number],
                'Product Name': [product_name],
                'Product Site': [product_site],
                'Review Generated By': [review_generated_by],
                'Reviews': [row[review]]
            })
            # Add the new row to the list
            rows_to_concat.append(new_row)
    
    # Concatenate all new rows into the transformed DataFrame
    if rows_to_concat:
        transformed_df = pd.concat([transformed_df] + rows_to_concat, ignore_index=True)

# Save the transformed DataFrame to a new CSV file
transformed_df.to_csv('~/Desktop/output.csv', index=False)

print("Transformation complete. Check the 'output.csv' file.")
```
